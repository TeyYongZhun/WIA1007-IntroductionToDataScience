```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.ibm.com/think/topics/artificial-intelligence'

page = requests.get(url)
```

```python
soup = BeautifulSoup(page.content, 'html.parser') #use.content to encode the text which is more safer than .text
```

```python
print(soup)
```

```python
soup.find_all('p')
```

```python
soup.find_all('p')[0].get_text()
```
